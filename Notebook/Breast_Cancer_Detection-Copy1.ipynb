# <font color="red"> k-Nearest Neighbors</font>

# Setting parameter ( Find best K )

````python
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# Senarai nilai k untuk diuji
k_values = range(1, 21)

# Simpan skor purata
mean_scores = []

for k in k_values:
    # Buat model KNN dengan nilai k tertentu
    knn = KNeighborsClassifier(n_neighbors=k)
    
    # Kira skor menggunakan cross-validation
    scores = cross_val_score(knn, x_train, y_train, cv=5, scoring='accuracy')
    
    # Simpan purata skor
    mean_scores.append(np.mean(scores))

# Paparkan nilai k terbaik
best_k = k_values[np.argmax(mean_scores)]
print(f"Nilai k terbaik ialah: {best_k}")

# Paparkan semua skor
for k, score in zip(k_values, mean_scores):
    print(f"k = {k},  average accuracy = {score:.3f}")
````

# Run Model

````python
# Import necessary libraries
from sklearn.neighbors import KNeighborsClassifier

# Initialize KNN with the best k value (n_neighbors)
optimized_knn_classifier = KNeighborsClassifier(
    n_neighbors= 5  # K terbaik yang diperoleh
)

# Train the model
optimized_knn_classifier.fit(x_train_scaled, y_train)
````

2) The most important metric depends on the context of the task and the cost of misclassifications:\

Accuracy: Useful if both classes are equally important and the dataset is balanced (as it is here).\
Precision: Important if the cost of false positives (Type I errors) is high, e.g., misclassifying non-critical cases as critical.\
Recall: Crucial if the cost of false negatives (Type II errors) is high, e.g., missing critical cases.\
F1-Score: A good balance between precision and recall, especially if both are equally important.\
Recommendation: Use F1-Score for its balance in evaluating overall performance, particularly when misclassification costs are similar.\
